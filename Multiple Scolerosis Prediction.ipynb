```python
def countplot(nn):
    #sns.set(rc={'figure.figsize':(12,8)})

    ax = sns.countplot(x=nn, palette = "pastel",
                       order=nn.value_counts().index.sort_values())

    abs_values = nn.value_counts()
    rel_values = nn.value_counts(normalize=True).values * 100
    lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]

    ax.bar_label(container=ax.containers[0], labels=lbls)
    plt.title("Variable Count Plot")
    plt.show()
```

```python
"""this section is to balance target variable"""
from sklearn.utils import resample

#assign major and minor class
df_majority = df[df.group== 2]
df_minority = df[df.group== 1]

#apply random resample based on major class count
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     
                                 n_samples=len(df_majority),    
                                 random_state=123) 

#concat the result into df
df = pd.concat([df_majority, df_minority_upsampled])
df = df.reset_index(drop = True)
countplot(df['group'])
```

```python

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score  
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report, roc_curve, auc
```

```python
#split the data to train and test with 80/20 ratio

x_train, x_test, y_train, y_test = train_test_split(df.drop(['group'], axis=1), 
                                                    df['group'],
                                                    test_size = 0.2,
                                                    shuffle =True,random_state =42)
```

```python
colist = [i for i in range(len(df.columns)+1)]
    
df.boxplot()
plt.title('Initial Symptom of Disease')
plt.xticks(ticks=colist, labels=colist)
plt.show();
```

```python
#assign scaler for train data
ss = StandardScaler()

#assign different train set for scaled data
x_train= pd.DataFrame(ss.fit_transform(x_train), columns = x_train.columns)
x_test= pd.DataFrame(ss.fit_transform(x_test), columns = x_train.columns)
```

```python
seed = 7
models = []
results = []
names = []

models.append(('LogReg', LogisticRegression()))
models.append(('RanFor', RandomForestClassifier()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('AdaBoost', AdaBoostClassifier()))
models.append(('GaussianNB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('GradBoost', GradientBoostingClassifier()))

scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
    cv_results = model_selection.cross_val_score(model, x_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
"""this function is to tune the parameter with randomized search cross validation"""

def partuning(x):
    rscv_fit = x.fit(x_train, y_train)
    best_parameters = rscv_fit.best_params_
    best_estimator = rscv_fit.best_estimator_
    print(best_parameters)
    print(best_estimator)
```

```python
"""this function is to display confusion matrix"""

def confmat(cm):
    ax = sns.heatmap(cm, annot=True, cmap='YlOrBr', fmt='g')
    ax.set_title('Confusion Matrix\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    ax.xaxis.set_ticklabels(categories)
    ax.yaxis.set_ticklabels(categories)

    plt.show()
```

```python
"""this function is to display evaluation of classification"""

def simpleeval(x,y,m):
    pred = m.predict(x)
    score=accuracy_score(y,pred)
    print('Accuracy : %.5f'%score)
    print(classification_report(y, pred, target_names = categories))
```

```python
def featureimp(columns, importances): 
    importances = pd.DataFrame(data={
        'Attribute': columns,
        'Importance': importances
    })
    importances = importances.sort_values(by='Importance', ascending=False)
    #plt.figure(figsize=(10, 10))
    plt.bar(x=importances['Attribute'][1:20], height=importances['Importance'][1:20], color='salmon')
    plt.title('Top 20 Impactful Feature', size=20)
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
#assign classifier
rft = RandomForestClassifier()

#assign tuning parameter
grid_rf = {
'n_estimators': [20, 50, 100, 500, 1000],  
'max_depth': np.arange(1, 15, 1),  
'min_samples_split': [2, 10, 9], 
'min_samples_leaf': np.arange(1, 15, 2, dtype=int),  
'bootstrap': [True, False], 
'random_state': [1, 2, 30, 42]
}

#assign RandomizedSearch cross validation with 5 iteration
rscv = RandomizedSearchCV(estimator = rft, 
                          param_distributions = grid_rf, 
                          cv = 10, 
                          n_jobs=-1, 
                          verbose=2, 
                          n_iter= 5)
```

```python
#call the function
partuning(rscv)
```

```python
"""this part is to fit the classifier to the data with best hyperparameter"""
#assign classifier
rdcl = RandomForestClassifier(max_depth=13, min_samples_split=9, random_state=30)
#fit the data
rdcl.fit(x_train, y_train)
#print the score of classifier
print("%.2f"%rdcl.score(x_test,y_test))
```

```python
#call the fucnction for random forest model evaluation
simpleeval(x_test,y_test,rdcl)
```

```python
#call the confusion matrix function 
pred = rdcl.predict(x_test) 
confm = confusion_matrix(y_test,pred)
confmat(confm)
```

```python
rfimp = rdcl.feature_importances_
xcolumn = x_train.columns
featureimp(xcolumn, rfimp)
```

```python
#assign classifier
gbt = GradientBoostingClassifier()

#assign tuning parameter
grid_gb = {
    "n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]
}


#assign RandomizedSearch cross validation with 5 iteration
rscv = RandomizedSearchCV(estimator = gbt, 
                          param_distributions = grid_gb, 
                          cv = 10, 
                          n_jobs=-1, 
                          verbose=2, 
                          n_iter= 5)
```

```python
#call the function
partuning(rscv)
```

```python
"""this part is to fit the classifier to the data with best hyperparameter"""
#assign classifier
gbcl = GradientBoostingClassifier(max_depth=5, n_estimators=50)
#fit the data
gbcl.fit(x_train, y_train)
#print the score of classifier
print("%.2f"%gbcl.score(x_test,y_test))
```

```python
#call the fucnction for random forest model evaluation
simpleeval(x_test,y_test,gbcl)
```

```python
#call the confusion matrix function 
pred = gbcl.predict(x_test) 
confm = confusion_matrix(y_test,pred)
confmat(confm)
```

```python
gbimp = gbcl.feature_importances_
xcolumn = x_train.columns
featureimp(xcolumn, gbimp)
```

```python
"""this section is to get and plot the ROC AUC"""

y_test0 = y_test-1

r_probs = [0 for _ in range(len(y_test0))]
rf_probs = rdcl.predict_proba(x_test)
gb_probs = gbcl.predict_proba(x_test)

rf_probs = rf_probs[:, 1]
gb_probs = gb_probs[:, 1]


r_auc = roc_auc_score(y_test0, r_probs)
rf_auc = roc_auc_score(y_test0, rf_probs)
gb_auc = roc_auc_score(y_test0, gb_probs)


r_fpr, r_tpr, _ = roc_curve(y_test0, r_probs)
rf_fpr, rf_tpr, _ = roc_curve(y_test0, rf_probs)
gb_fpr, gb_tpr, _ = roc_curve(y_test0, gb_probs)

plt.plot(r_fpr, r_tpr, linestyle='--')
plt.plot(rf_fpr, rf_tpr, linestyle='-', label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(gb_fpr, gb_tpr, linestyle='-', label='Gradient Boosting (AUROC = %0.3f)' % gb_auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() 
# Show plot
plt.show()
```
